## Librerías

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./../Datasets/full_RASFF_DATA.csv', sep=';', header=0, index_col = 0)
df.head()
```

```python
display(len(df))
display(len(df['HAZARDS_CAT'].unique()))
display(len(df['HAZARDS'].unique()))
display(len(df['PROD_CAT'].unique()))
display(len(df['PRODUCT'].unique()))
```

### Obtención de información del dataset

#### Df_Productos

```python
#Lista con las repeticiones de los productos y categorías
lista_productos = df['PRODUCT'].value_counts()
df_productos = pd.DataFrame(data = {'Producto': lista_productos.index, 'Repeticiones': lista_productos.values})
df_productos['%_Aparicion_dataset'] = df_productos['Repeticiones']/df.shape[0]

```

#### Df_Cat_Productos

```python
#Lista con las repeticiones de las categorías de los productos
lista_cat_productos = df['PROD_CAT'].value_counts()
df_cat_productos = pd.DataFrame(data = {'Cat_Producto': lista_cat_productos.index, 'Repeticiones': lista_cat_productos.values})
df_cat_productos['%_Aparicion_dataset'] = df_cat_productos['Repeticiones']/df.shape[0]
```

#### Df_Amenazas

```python
#Lista con las repeticiones de las amenazas 
lista_amenazas = df['HAZARDS'].value_counts(dropna = False)
df_amenazas = pd.DataFrame(data = {'Amenaza': lista_amenazas.index, 'Repeticiones': lista_amenazas.values})
df_amenazas['%_Aparicion_dataset'] = df_amenazas['Repeticiones']/df.shape[0]
```

#### Df_Cat_Amenaza_SinTratar

```python
#Aquí se muestran las amenazas sin tratar, hay amenazas múltiples por campo, siendo un total de 552.

lista_catHazard =df['HAZARDS_CAT'].value_counts()
df_catHaz = pd.DataFrame(data = {'Cat_Amenaza': lista_catHazard.index, 'Repeticiones': lista_catHazard.values})
df_catHaz['%_Aparicion_dataset'] = df_catHaz['Repeticiones']/df.shape[0]
#df_catHaz
```

```python
df_catHaz[75:].sum()
```

#### Df_Cat_Amenaza

```python
#Lista con las repeticiones de las categorías de amenazas en bruto
lista_cat_amenazas = []

for i in range(df.shape[0]):
    if(not pd.isna(df.iloc[i,12])):
        lista_cat_amenazas.append(df.iloc[i,12].split(",")[0]) 
```

```python
#Lista con las categorías de amenazas (tratado)
#Tratamos los datos para obtener las categorías de las amenazas
repeticiones_cat_amenazas = {'Cat_Amenaza':[i for i in lista_cat_amenazas],'Repeticiones':[lista_cat_amenazas.count(i) for i in lista_cat_amenazas]}

df_cat_amenazas = pd.DataFrame.from_dict(repeticiones_cat_amenazas)
df_cat_amenazas = df_cat_amenazas.drop_duplicates().reset_index().drop(['index'], axis = 1)
df_cat_amenazas['%_Aparicion_dataset'] = df_cat_amenazas['Repeticiones']/df.shape[0]
df_cat_amenazas = df_cat_amenazas.sort_values('Repeticiones', ascending = False)
```

#### Df_Paises

```python
#Lista con las repeticiones de los países en bruto
lista_paises = []

for i in range(df.shape[0]):
    if(not pd.isna(df.iloc[i,13])):
        lista_paises = lista_paises + df.iloc[i,13].split(",")
    if(not pd.isna(df.iloc[i,14])):
        lista_paises = lista_paises + df.iloc[i,14].split(",")
    if(not pd.isna(df.iloc[i,15])):
        lista_paises = lista_paises + df.iloc[i,15].split(",")
```

```python
#Lista con las repeticiones de los países (tratado)
#Tratamos los datos para obtener la infromación acerca de los paises.
repeticiones_paises = {'Pais':[i for i in lista_paises],'Repeticiones':[lista_paises.count(i) for i in lista_paises]}

df_repes = pd.DataFrame.from_dict(repeticiones_paises)
df_repes = df_repes.drop_duplicates().reset_index().drop(['index'], axis = 1)
df_repes['%_Aparicion_dataset'] = df_repes['Repeticiones']/df.shape[0]
df_repes = df_repes.sort_values('Repeticiones', ascending = False)
```

#### Guardamos la información obtenida en CSVs.

```python
df_productos.to_csv('./../Datasets/Lista_Productos.csv', index=True)
df_cat_productos.to_csv('./../Datasets/Lista_Categoria_Productos.csv', index=True)
df_amenazas.to_csv('./../Datasets/Lista_Amenazas.csv', index=True)
df_cat_amenazas.to_csv('./../Datasets/Lista_Categoria_Amenazas.csv', index=True)
df_repes.to_csv('./../Datasets/Lista_repeticion_paises.csv', index=True)
```
