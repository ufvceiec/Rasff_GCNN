```python
import numpy as np
import pandas as pd
```

```python
#Viejo dataset
data = pd.read_csv('./full_RASFF_DATA_16092019.csv', sep=';', 
                   header=0, index_col = 0)
old = data
```

```python
#Nuevo dataset
data = pd.read_csv('./new_RASFF_DATA.csv', sep='\t', 
                   header=0, index_col = 0)
new = data
```

```python
#Uni√≥n de ambos datasets
df = pd.concat([new,old])
```

```python
#Tratamiento de formato del nuevo dataset.

#Borramos duplicados
df = df.drop_duplicates(subset='REF')

#Eliminamos registros no v√°lidos.
df = df.dropna(how='all')
df = df.dropna(subset = ['DATE_CASE'])
data = data.fillna(" ")

#Ordenar los registros por fecha .
df['DATE_CASE'] = pd.to_datetime(df['DATE_CASE'],  errors='coerce', dayfirst=True, format = '%d/%m/%Y')
df= df.sort_values(by='DATE_CASE',ascending=False)
df = df.reset_index(drop = True)
#df = df.rename(columns={"COUNT_ORIGEN": "ORIG_COUNTRY", "COUNT_DESTIN": "DESTIN_COUNTRY", "COUNT_CONCERN": "CONCERN_COUNTRY"})
df = df[df.CLASSIF != ' ']

#Poner estos campos en lowercase.
df ['CLASSIF'] = df ['CLASSIF'].str.lower()
df ['PROD_CAT'] = df ['PROD_CAT'].str.lower()
df ['TYPE'] = df ['TYPE'].str.lower()
df ['RISK_DECISION'] = df ['RISK_DECISION'].str.lower()
df ['ACTION_TAKEN'] = df ['ACTION_TAKEN'].str.lower()
df ['DISTRIBUTION_STAT'] = df ['DISTRIBUTION_STAT'].str.lower()
df ['PRODUCT'] = df ['PRODUCT'].str.lower()
df ['HAZARDS'] = df ['HAZARDS'].str.lower()
df ['HAZARDS_CAT'] = df ['HAZARDS_CAT'].str.lower()

#Debug
display(df)
print(df.dtypes)
```

```python
#Dataset que genera.
df.to_csv('./full_RASFF_DATA.csv', sep=';')
```
