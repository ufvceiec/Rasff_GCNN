```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./full_RASFF_DATA.csv', sep=';', header=0, index_col = 0)
```

```python
#Eliminamos las fechas que no nos interesan.
fecha_maxima = "2021" #Primer año que no queremos coger.
df = df.loc[df['DATE_CASE'] < fecha_maxima]
```

```python
df[0:10]
```

```python
numRegistros = np.zeros((42))
numTipoRegistros = np.zeros((4)) #alert, border rejection, information for attention o information for follow up. En 2020
```

```python
dict_classif = {
    "alert": 0,
    "border rejection": 1,
    "information for attention": 2,
    "information for follow-up": 3
}
```

```python
contador = 0
fecha_anterior = "2020"

for index, row in df.iterrows():
    #Asignamos
    if(index == 0):
        fecha_anterior = row['DATE_CASE'].split('-')[0]
    #Miramos
    if(row['DATE_CASE'].split('-')[0] != fecha_anterior):
        contador += 1
        fecha_anterior = row['DATE_CASE'].split('-')[0]

    numRegistros[contador] += 1
    #2020
    if((row['DATE_CASE']<"2021") & (row['DATE_CASE']>"2019")):
        numTipoRegistros[dict_classif.get(row['CLASSIF'])] += 1 
        
```

```python
fechas = np.arange(2020,1978,-1)
#fechas = ["2020","2019","2018","2017","2016","2015","2014","2013","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020",
          #"2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020","2020",]
#fechas = np.array(map(str, fechas))
#map(str, fechas)
#fechas = fechas.astype("|S7")
#display(fechas)
```

```python
display(numTipoRegistros)
display(numRegistros)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
labels = fechas
values = numRegistros
ax.bar(labels,values)
plt.xticks(rotation=90)
ax.set_xlabel("DATE CASE")
plt.savefig('./DATE_CASE_CASES.png')
plt.show()
```
